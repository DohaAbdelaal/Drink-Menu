```python
#imports Libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import dataset
df = pd.read_csv('drinkMenu.csv')
print(df.info()) 
print (df.shape)
df.isna().sum()
```

```python
#remove duplicates
df.drop_duplicates(subset=None, keep="first", inplace=True)
```

```python
#Replace nulls
df.fillna(df['Caffeine (mg)'].mode()[0], inplace = True)
```

```python
#Data visualization

#Pull the tow columns from the dataset and rename them
df.calories = df[['Beverage_category','Calories']]
df.calories.columns = ['DrinkMenu', 'Calories']
df.calories
sns.countplot(y='DrinkMenu', data=df.calories)
```

```python
#To see which Starbucks drink has the highes calories Beverage_category
df.calories.sort_values("Calories", ascending=False)
```

```python
# data visualization to see the which Starbucks drink has the highest calories
plt.figure(figsize=(15, 8))
sns.barplot(x="Beverage", y="Calories", data=df)
plt.xticks(rotation=40, ha='right')
plt.title(" Beverage Calories")
plt.show()
```

```python
#highest sugar drink at sturbucks
df.sort_values(" Sugars (g)", ascending=False)
```

```python
# data visualization to see the which Starbucks has the highest sugar amount
plt.figure(figsize=(15, 8))
sns.barplot(x="Beverage", y=" Sugars (g)", data=df)
plt.xticks(rotation=40, ha='right')
plt.title(" Beverage Sugar(g)")
plt.show()
```
